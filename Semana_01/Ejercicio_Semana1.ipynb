```python
from google.colab.patches import cv2_imshow

im2 = np.zeros((200, 200, 3), np.uint8)

colors = [(0, 0, 255), (0, 255, 0), (255, 0, 0), (0, 0, 0)]
start, end = 50, 150

for color in colors:
    cv2.rectangle(im2, (start, start), (end, end), color, -1)
    start += 10
    end -= 10
im2
```

```python
im3 = np.zeros((200, 200), np.uint8)

# Dibujar un patr√≥n de puntos
for i in range(0, 200, 10):
    for j in range(0, 200, 10):
        if (i + j) % 20 == 0:
            cv2.circle(im3, (i, j), 3, (255, 255, 255), -2)
        else:
            cv2.circle(im3, (i, j), 3, (128, 128, 128), -2)

im3
```

```python
im = np.zeros((200, 200, 3), dtype=np.uint8)

for i in range(200):
    for j in range(200):
        im[i, j] = (255 * (0.5 * np.sin(i * 0.1) + 0.5 * np.sin(j * 0.1)),
                    255 * (0.5 * np.sin(i * 0.1) + 0.5 * np.sin(j * 0.1)),
                    255 * (0.5 * np.sin(i * 0.1) + 0.5 * np.sin(j * 0.1)))

im
```
